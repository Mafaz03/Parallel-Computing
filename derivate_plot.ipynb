```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("derivate.dat")
```

```python
num = data[:, 0]   
exact = data[:, 1] 


plt.plot(num, label="Numerical")
plt.plot(exact, label="Exact")

plt.legend()
plt.xlabel("Index")
plt.ylabel("Derivative")
plt.title("Numerical vs Exact Derivative")
plt.show()
```
